### Decision Tree Visualization

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

# Create DecisionTree Classifier
dt_clf = DecisionTreeClassifier(random_state=156)

# Load Iris data, and then split training set and test set
iris_data = load_iris()
X_train , X_test , y_train , y_test = train_test_split(iris_data.data, iris_data.target,
                                                       test_size=0.2,  random_state=11)

# Train DecisionTreeClassifer. 
dt_clf.fit(X_train , y_train)
```

```python
from sklearn.tree import export_graphviz

# Create tree.dot file(out_file) from the result of calling export_graphviz.  
export_graphviz(dt_clf, out_file="tree.dot", class_names=iris_data.target_names , \
feature_names = iris_data.feature_names, impurity=True, filled=True)
```

```python
import graphviz

# read tree.dot file from Graphviz, and visualize
with open("tree.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

### feature importance
- a technique used to determine the relative significance of input variables in making predictions. 

>  It helps identify crucial features for better model performance, feature selection, and model interpretability.

```python
import seaborn as sns
import numpy as np
%matplotlib inline

# extract feature importance 
print("Feature importances:\n{0}".format(np.round(dt_clf.feature_importances_, 3)))

# mapping importance by feature
for name, value in zip(iris_data.feature_names , dt_clf.feature_importances_):
    print('{0} : {1:.3f}'.format(name, value))

# visualize feature importance by column 
sns.barplot(x=dt_clf.feature_importances_ , y=iris_data.feature_names)
```

### Decision Tree & Overfitting

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("3 Class values with 2 Features Sample data creation")

# For 2D visualization, the number of feature = 2, and result class created 3 types of classification sample data set. 
X_features, y_labels = make_classification(n_features=2, n_redundant=0, n_informative=2,
                             n_classes=3, n_clusters_per_class=1,random_state=0)

# visualization of 2 features in plot type, each class value is represented in different color.
plt.scatter(X_features[:, 0], X_features[:, 1], marker='o', c=y_labels, s=25, cmap='rainbow', edgecolor='k')
```

```python
import numpy as np

# function : visualize Decision Boundary of Classifier
def visualize_boundary(model, X, y):
    fig,ax = plt.subplots()
    
    # scatter plot for training data
    ax.scatter(X[:, 0], X[:, 1], c=y, s=25, cmap='rainbow', edgecolor='k',
               clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim_start , xlim_end = ax.get_xlim()
    ylim_start , ylim_end = ax.get_ylim()
    
    model.fit(X, y)

    xx, yy = np.meshgrid(np.linspace(xlim_start,xlim_end, num=200),np.linspace(ylim_start,ylim_end, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    

    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap='rainbow', clim=(y.min(), y.max()),
                           zorder=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

# visualize Decsion Boundary 
dt_clf = DecisionTreeClassifier(random_state=156).fit(X_features, y_labels)
visualize_boundary(dt_clf, X_features, y_labels)
```

```python
# visualize Decsion Boundary  with min_samples_leaf=6 
dt_clf = DecisionTreeClassifier(random_state=156, min_samples_leaf=6).fit(X_features, y_labels)
visualize_boundary(dt_clf, X_features, y_labels)
```
