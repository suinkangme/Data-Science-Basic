```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        result.append(float(row[-1]))
        
print(result)
```

```python
print(len(result))
```

```python
import matplotlib.pyplot as plt
plt.plot(result,'r')
plt.figure(figsize=(10,2))
plt.show()
```

```python
for row in data:
    if row[-1] != '':
        if row[0].split('_')[1]=='07' and row[0].split('-')[2]=='05':
            result.append(float(row[-1]))

plt.plot(result, 'hotpink')
plt.title('the temperature change graph for my birthday')
plt.show()
```
